```r
drds <- "./03FN_RDS/"
```

```r
gene <- c("C1qa","C1qb","C1qc","Trem2","Tmem119","Hexb","Ms4a7","Ms4a6c")
pdf("fig3_Feature.pdf",width = 12,height = 12)
for(s in Sys.glob(file.path(drds,"*.RDS"))){
    key <- gsub(".RDS","",basename(s))
    sp <- readRDS(paste0(drds,key,".RDS"))
    for(i in gene){
        p <- SpatialFeaturePlot(sp,features = i,pt.size.factor = 0.25,min.cutoff = 0.1)+ggplot2::scale_fill_continuous(low ="#54BAD3",high = "#ce1256")+coord_cartesian(xlim = c(0, 22000), ylim = c(0,22000))+labs(subtitle = key)
        print(p)
    }
}
dev.off()
```
