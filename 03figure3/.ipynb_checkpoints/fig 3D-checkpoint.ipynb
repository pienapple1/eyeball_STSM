```r
sp <- readRDS("/home/ljpuser/disk1/pienapple/03eye/30data_st_del/02data_del_nor/08RDS_mer_red/merge_red.RDS")
```

```r
####deg num
sp@meta.data$deg_num <- 0
sp@meta.data[which(sp@meta.data$area=="CB"),]$deg_num <- 153
sp@meta.data[which(sp@meta.data$area=="Cornea&conj"),]$deg_num <- 258
sp@meta.data[which(sp@meta.data$area=="GCL"),]$deg_num <- 731
sp@meta.data[which(sp@meta.data$area=="INL"),]$deg_num <- 251
sp@meta.data[which(sp@meta.data$area=="Iris"),]$deg_num <- 126
sp@meta.data[which(sp@meta.data$area=="Len"),]$deg_num <- 346
sp@meta.data[which(sp@meta.data$area=="Muscle"),]$deg_num <- 18
sp@meta.data[which(sp@meta.data$area=="ONL"),]$deg_num <- 116
sp@meta.data[which(sp@meta.data$area=="Optic nerve"),]$deg_num <- 205
sp@meta.data[which(sp@meta.data$area=="RPE"),]$deg_num <- 22
```

```r
table(sp@meta.data$deg_num)
```

```r
pdf("Deg_Feature.pdf",width = 12,height = 12)
for(i in unique(sp@meta.data$sample)){
    sp1 <-subset(sp,subset=sample==i)
    coord.df = data.frame(x=sp1@meta.data$x, y=sp1@meta.data$y, stringsAsFactors=FALSE)
    rownames(coord.df) = rownames(sp1@meta.data)
    sp1@images$image = new(
    Class = 'SlideSeq',
    assay = "RNA",
    key = "image",
    coordinates = coord.df)
    p <- SpatialFeaturePlot(sp1,features = "deg_num",pt.size.factor = 0.5)+coord_cartesian(xlim = c(0, 22000), ylim = c(0,22000))+labs(subtitle = i)
    print(p)
}
dev.off()
```
