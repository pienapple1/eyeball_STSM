```r
pdf("eye_st_area_divi.pdf",width = 6,height = 6)
for(s in Sys.glob(file.path(drds,"*.RDS"))){
    key <- gsub(".RDS","",basename(s))
    sp <- readRDS(paste0(drds,key,".RDS"))
    p <- SpatialPlot(sp,group.by = "area",cols = cols2,pt.size.factor = 1.2)+labs(subtitle = key)+theme(legend.position = "none")
    print(p)
    p <- SpatialFeaturePlot(sp,features = "layer",pt.size.factor = 1.2)+labs(subtitle = key)+theme(legend.position = "none")
    print(p)
}
dev.off()
```
