```r
pexp <- read.csv("eye_protein.csv",row.names = 1)
pexp
```

```r
m <- pexp
```

```r
m$group <- substr(rownames(m),1,3)
```

```r
m$group <- str_replace(m$group,"Cat","Control")
m$group <- str_replace(m$group,"Gla","POAG")
```

```r
unique(m$group)
```

```r
colnames(pexp)
```

```r
col<-c("Control"="#337AB7","POAG"="#F0AD4E")
```

```r
i="AHCY"
p <- ggplot(m,aes(x=group,y=m[[i]],fill=group))+geom_boxplot(position=position_dodge(width=0.2), width=0.4) +
stat_boxplot(geom="errorbar", position=position_dodge(width=0.4), width=0.2)+#geom_point(aes(fill=group, group=group, color=group), pch=21, size=0,position=position_dodge(0.4)) +
geom_signif(comparisons = list(c('Control','POAG')),map_signif_level = T,test = "wilcox.test",step_increase = 0.05)+
scale_fill_simpsons(alpha=0.7)+scale_color_simpsons(alpha=0.7)+
labs(x=NULL, y=NULL,subtitle = i)+
theme(
    panel.border = element_rect(
      color = "black",
      fill = NA,
      linewidth = 1
    ),panel.background = element_rect(fill = "white")
  )+
theme(axis.text.y=element_text(color="black", size=10, face="bold"), # 自定义y轴上的文本
axis.text.x=element_text(color="black", size=10, hjust=1, face="bold", angle=45) # 自定义x轴上的文本
     )+scale_fill_manual(values = col) +scale_color_manual(values = col)
print(p)
```
