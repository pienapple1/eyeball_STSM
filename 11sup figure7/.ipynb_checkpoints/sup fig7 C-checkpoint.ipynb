```r
m1 <- read.csv("metabolism_C16.csv",row.names = 1,check.names = FALSE)
```

```r
####limma deg
library(dplyr) # 数据处理
library(data.table) # 数据读取
library(stringr) # 数据处理
library(limma) # limma差异分析
library(edgeR) # edgeR差异分析
library(DESeq2) # DESeq2差异分析 
```

```r
df <- m1
df$group <- substr(rownames(df),1,4)
df$group <- str_replace(df$group,"GLAU","POAG")
df
```

```r
# 生成分组信息，处理组在前，对照组在后
comp <- unlist(strsplit("POAG_vs_CTRL",split = "_vs_"))  
group_list <- factor(df$group,levels = comp)  # factor因子用于存储不同类别的数据，可以用来对数据进行分组，levels是指定各水平值,不指定默认按字母排序
group_list    # 查看分组信息
table(group_list)  
```

```r
# 构建线性模型，0代表x线性模型的截距为0
design <- model.matrix(~0+group_list)  # 生成设计矩阵，又叫分组矩阵，分组依据是factor的level水平
rownames(design) <- colnames(t(m1))  # 设计矩阵的行名，与表达矩阵的列名一一对应
colnames(design) <- levels(factor(group_list))   # 设计矩阵的列名，与生成的分组信息一致
design
```

```r
m <- df
```

```r
col<-c("CTRL"="#337AB7","POAG"="#F0AD4E")
```

```r
i="SM C16:0"
p <- ggplot(m,aes(x=group,y=m[[i]],fill=group))+geom_boxplot(position=position_dodge(width=0.2), width=0.4,alpha=0.1) +
stat_boxplot(geom="errorbar", position=position_dodge(width=0.4), width=0.2)+geom_point(aes(fill=group, group=group, color=group), pch=21, size=3,position=position_dodge(0.4)) +
geom_signif(comparisons = list(c('CTRL','POAG')),map_signif_level = T,test = "wilcox.test",step_increase = 0.05)+
scale_fill_simpsons(alpha=0.7)+scale_color_simpsons(alpha=0.7)+
labs(x=NULL, y=NULL,subtitle = i)+
theme(
    panel.border = element_rect(
      color = "black",
      fill = NA,
      linewidth = 1
    ),panel.background = element_rect(fill = "white")
  )+
theme(axis.text.y=element_text(color="black", size=10, face="bold"), # 自定义y轴上的文本
axis.text.x=element_text(color="black", size=10, hjust=1, face="bold", angle=45) # 自定义x轴上的文本
     )+scale_fill_manual(values = col) +scale_color_manual(values = col)
print(p)
```

```r
pdf("./human_metabolism_C16.pdf",width = 4,height = 4.5)
for(i in colnames(m1)){
    p <- ggplot(m,aes(x=group,y=m[[i]],fill=group))+geom_boxplot(position=position_dodge(width=0.2), width=0.4,alpha=0.1) +
stat_boxplot(geom="errorbar", position=position_dodge(width=0.4), width=0.2)+geom_point(aes(fill=group, group=group, color=group), pch=21, size=3,position=position_dodge(0.4)) +
geom_signif(comparisons = list(c('CTRL','POAG')),map_signif_level = T,test = "wilcox.test",step_increase = 0.05)+
scale_fill_simpsons(alpha=0.7)+scale_color_simpsons(alpha=0.7)+
labs(x=NULL, y=NULL,subtitle = i)+
theme(
    panel.border = element_rect(
      color = "black",
      fill = NA,
      linewidth = 1
    ),panel.background = element_rect(fill = "white")
  )+
theme(axis.text.y=element_text(color="black", size=10, face="bold"), # 自定义y轴上的文本
axis.text.x=element_text(color="black", size=10, hjust=1, face="bold", angle=45) # 自定义x轴上的文本
     )+scale_fill_manual(values = col) +scale_color_manual(values = col)
print(p)
}
dev.off()
```
