```r
mer <- subset(sc@meta.data,subset=area!="Optic nerve")
```

```r
cell_order <- c(
  # 基质细胞
  'Fibroblast', 'VEC', 'CEC', 'LEC', 'SMC', 'Muscle', 'RPE', 
  'Conj', 'LSC', 'CSK', 'PCE', 'NPCE', 'Proliferation cell',
  
  # 神经细胞
  'Rod', 'Cone', 'RGC', 'AC', 'HC', 'CBC', 'RBC','Macroglia', 
   'Oligodendrocyte', 'Schwann cell', 'Melanocyte',
  # 免疫细胞
  'Microglia','Myeloid', 'Neutrophil', 'DC', 'Lymphocyte'
)
```

```r
head(mer)
mer$dgdivi <- str_replace(mer$dgdivi,"Optic nerve","Optic")
mer$groupA <- paste0(mer$group,"_",mer$dgdivi)
unique(mer$groupA)
mer$groupA <- factor(mer$groupA,levels = c("Sham_1","RIR_1","Sham_2","RIR_2"))
mer$celltype <- factor(mer$celltype,levels = cell_order)
```
