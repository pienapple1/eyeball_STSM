```r
drds <- "./07FN_RDS/"
n=0
for(s in Sys.glob(file.path(drds,"*_fn.RDS"))){
    n=n+1
    key <- gsub("_fn.RDS","",basename(s))
    sp <- readRDS(paste0(drds,key,"_fn.RDS"))
    meta <- sp@meta.data
    rownames(meta) <- paste0(key,rownames(meta))
    meta$chip <- key
    counts <- as.data.frame(sp@assays$RNA$counts)
    colnames(counts) <- paste0(key,colnames(counts))
    if(n==1){
        meta1 <- meta
        counts1 <- counts
    }
    else{
        meta1 <- rbind(meta1,meta)
        counts1 <- merge(counts1,counts,all=TRUE,by=0)
        rownames(counts1) <- counts1$Row.names
        counts1 <- counts1[,-1]
    }
}
```

```r
library(edgeR)

# edgeR的cpm函数使用
cpm_edgeR <- function(counts_matrix, log = FALSE, prior.count = 0.25) {
  # log: 是否取log2转换
  # prior.count: prior count用于避免对0取log
  return(cpm(counts_matrix, log = log, prior.count = prior.count))
}

# 使用示例：
# cpm_counts <- cpm_edgeR(counts)
library(DESeq2)

deseq_cpm_like <- function(counts_matrix) {
  dds <- DESeqDataSetFromMatrix(countData = counts_matrix,
                               colData = data.frame(sample=colnames(counts_matrix)),
                               design = ~1)
  
  # Estimate size factors for normalization
  dds <- estimateSizeFactors(dds)
  
  # Get normalized counts (similar to CPM but using size factors instead of total counts)
  norm_counts <- counts(dds, normalized=TRUE)
  
  # Scale to counts per million
  norm_counts_cpm <- norm_counts * (1e6 / colSums(norm_counts))
  
  return(norm_counts_cpm)
}
```

```r
log2cpm_plus1 <- log2(deseq_cpm_like(counts1) + 1)
```
