```r
pbmc <- seurat
DefaultAssay(pbmc) <- "RNA"
pbmc <- NormalizeData(pbmc,normalization.method = "LogNormalize")
Idents(pbmc) <- as.factor(pbmc@meta.data$celltype)
```

```r
#选择前2000个高变基因
pbmc <- FindVariableFeatures(pbmc, selection.method = "vst", nfeatures = 3000)
#选择前10个基因添加数据标签
top10 <- head(VariableFeatures(pbmc), 10)
p1 <- VariableFeaturePlot(pbmc)
p2 <- LabelPoints(plot = p1, points = top10, repel = TRUE)
p1
p2
```

```r
all.genes <- rownames(pbmc)
pbmc <- ScaleData(pbmc, features = all.genes)
pbmc <- RunPCA(pbmc, features = VariableFeatures(pbmc),npcs = 120)#####select deg unique(markers$gene) fea
```

```r
#按照样本去批次，也可以选择根据group
library(harmony)
sce2 = pbmc %>% RunHarmony("sample", plot_convergence = TRUE)
#标准流程，参数不变
sce2 <- sce2 %>% 
  RunUMAP(reduction = "harmony", dims = 1:100,n.neighbors=80,min.dist=0.8,n.components=2) %>% 
  RunTSNE(reduction = "harmony", dims = 1:100,n.neighbors=80,min.dist=0.8,n.components=2) %>% 
  FindNeighbors(reduction = "harmony", dims = 1:100) %>% 
  FindClusters(resolution = 3) %>% 
  identity()
```

```r
DimPlot(sce2, reduction = "umap", pt.size=0.5, label = TRUE,repel = TRUE,cols = color)
DimPlot(sce2, reduction = "tsne", pt.size=0.5, label = TRUE,repel = TRUE,cols = color)
DimPlot(sce2, reduction = "tsne",group.by = "celltype", pt.size=0.5, label = TRUE,repel = TRUE,cols = color)
DimPlot(sce2, reduction = "umap",group.by = "celltype", pt.size=0.5, label = TRUE,repel = TRUE,cols = color)
```
